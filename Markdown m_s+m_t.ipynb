Assume you define the discrete Fourier transform on \(\Z_k^n\) by  

$\hat f(\xi)=\sum_{x\in\Z_k^n}f(x)\,e^{-2\pi i\langle x,\xi\rangle/k}$,

and inversion by  

$f(x)=\frac1{k^n}\sum_{\xi\in\Z_k^n}\hat f(\xi)\,e^{2\pi i\langle x,\xi\rangle/k}$.

Suppose moreover that $(\hat f(\xi)=0)$ for all $(\xi\notin S\subset\Z_k^n)$.  Then:

---

### 1. Fourier inversion + triangle inequality


\begin{aligned}
f(x)
&=\frac1{k^n}\sum_{\xi\in S}\hat f(\xi)\,e^{2\pi i\langle x,\xi\rangle/k}
\;\;\Longrightarrow\;\;
|f(x)|\;\le\;\frac1{k^n}\sum_{\xi\in S}\bigl|\hat f(\xi)\bigr|.
\end{aligned}


---

### 2. Cauchy–Schwarz on the finite sum

By Cauchy–Schwarz,
$
\sum_{\xi\in S}\bigl|\hat f(\xi)\bigr|
\;\le\;\sqrt{\sum_{\xi\in S}1}\;\sqrt{\sum_{\xi\in S}\bigl|\hat f(\xi)\bigr|^2}
\;=\;\sqrt{|S|}\;\|\hat f\|_{\ell^2(\Z_k^n)},
$
so
$
|f(x)|\;\le\;\frac{\sqrt{|S|}}{k^n}\,\|\hat f\|_{\ell^2(\Z_k^n)}.
$

---

### 3. Plancherel identity (no normalization in the forward transform)

One shows
$
\sum_{x\in\Z_k^n}|f(x)|^2
\;=\;\frac1{k^n}\sum_{\xi\in\Z_k^n}|\hat f(\xi)|^2
\quad\Longrightarrow\quad
\|\hat f\|_{\ell^2(\Z_k^n)} \;=\; k^{n/2}\,\|f\|_{\ell^2(\Z_k^n)}.
$
Substitute into the previous bound:
$
|f(x)|
\;\le\;\frac{\sqrt{|S|}}{k^n}\,\bigl(k^{n/2}\|f\|_{\ell^2}\bigr)
\;=\;\sqrt{|S|}\;k^{-n/2}\,\|f\|_{\ell^2}.
$

---

### 4. Conclusion

Since the above holds for every \(x\in\Z_k^n\),  
$
\boxed{\quad
\|f\|_{\ell^\infty(\Z_k^n)}
\;\le\;\frac{\sqrt{|S|}}{k^{n/2}}\;\|f\|_{\ell^2(\Z_k^n)},
}
$
i.e. you may take  
$
c \;=\;\frac{\sqrt{|S|}}{k^{n/2}}.
$
